# 实现Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1., 2., 3., 4., 5.])
y = np.array([1., 3., 2., 3., 5.])
```

```python
plt.scatter(x, y)
plt.axis([0, 6, 0, 6])
```

```python
from IPython.display import Image
Image(filename=r"image/x6.png")
```

```python
# 均值
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
# 分子
num = 0.0
# 分母
d = 0.0

# 求和
for x_i, y_i in zip(x, y):
    num += (x_i - x_mean) * (y_i - y_mean)
    d += (x_i - x_mean) ** 2
```

```python
a = num / d
b = y_mean - a * x_mean
```

```python
a
```

```python
b
```

```python
y_hat = a * x + b
```

```python
plt.scatter(x, y)
plt.plot(x, y_hat, color="r")
plt.axis([0, 6, 0, 6])
```

```python
x_predict = 6
y_predict = a * x_predict + b
```

```python
y_predict
```
